## **Patch: Robust Dataset Loading**

This cell modifies `train_segmentation.py` to correctly handle both `load_from_disk()` (local HuggingFace datasets) and `load_dataset()` sources, ensuring compatibility with saved datasets and preventing split-related errors.

```python
from pathlib import Path
import os

fp = Path("/kaggle/working/diarizers/train_segmentation.py")
lines = fp.read_text().splitlines(True)  # keep newline chars


for i, l in enumerate(lines):
    if l.strip().startswith("from datasets import") and "load_dataset" in l:
        # Make sure we have load_from_disk + Dataset + DatasetDict
        lines[i] = "from datasets import load_dataset, load_from_disk, Dataset, DatasetDict\n"
        break
else:
    raise RuntimeError("Couldn't find datasets import line to patch.")


start = None
end = None
for i, l in enumerate(lines):
    if "# Load the Dataset" in l:
        start = i
        break
if start is None:
    raise RuntimeError("Couldn't find '# Load the Dataset' section.")

for j in range(start + 1, len(lines)):
    if "train_split_name" in lines[j]:
        end = j
        break
if end is None:
    raise RuntimeError("Couldn't find the end of the dataset load block (train_split_name line).")

new_block = [
"    # Load the Dataset:\n",
"    dataset_name = str(data_args.dataset_name)\n",
"    if os.path.isdir(dataset_name) and (\n",
"        os.path.exists(os.path.join(dataset_name, 'dataset_info.json'))\n",
"        or os.path.exists(os.path.join(dataset_name, 'state.json'))\n",
"    ):\n",
"        dataset = load_from_disk(dataset_name)\n",
"        # If it's a single Dataset (no splits), wrap it so the rest of the script works\n",
"        if isinstance(dataset, Dataset):\n",
"            dataset = DatasetDict({'train': dataset})\n",
"    elif data_args.dataset_config_name:\n",
"        dataset = load_dataset(\n",
"            str(data_args.dataset_name),\n",
"            str(data_args.dataset_config_name),\n",
"            num_proc=int(data_args.preprocessing_num_workers)\n",
"        )\n",
"    else:\n",
"        dataset = load_dataset(\n",
"            str(data_args.dataset_name),\n",
"            num_proc=int(data_args.preprocessing_num_workers)\n",
"        )\n",
"\n",
]

lines[start:end] = new_block
fp.write_text("".join(lines))
print("✅ Patched:", fp)

print("\n--- Patched section preview ---")
for k in range(start, min(start + 35, len(lines))):
    print(f"{k+1:4d}: {lines[k].rstrip()}")
```

## **Build HuggingFace Dataset for Diarization Training**

This cell converts the raw audio files and CSV annotations into a structured HuggingFace `Dataset`, creates train and validation splits, and saves it to disk for use with the diarization training script.

```python
import os
import pandas as pd
from datasets import Dataset, DatasetDict, Audio

audio_dir = "/kaggle/input/competitions/dl-sprint-4-0-bengali-speaker-diarization-challenge/diarization/diarization/train/audio"
ann_dir   = "/kaggle/input/competitions/dl-sprint-4-0-bengali-speaker-diarization-challenge/diarization/diarization/train/annotation"

def time_to_seconds(t: str) -> float:
    # supports "HH:MM:SS" or "HH:MM:SS.mmm"
    h, m, s = t.strip().split(":")
    return int(h)*3600 + int(m)*60 + float(s)

rows = []
for wav in os.listdir(audio_dir):
    if not wav.endswith(".wav"):
        continue

    rec = wav[:-4]
    csv_path = os.path.join(ann_dir, rec + ".csv")
    if not os.path.exists(csv_path):
        continue

    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip().str.lower()

    rows.append({
        "audio": os.path.join(audio_dir, wav),
        "timestamps_start": [time_to_seconds(x) for x in df["start_time"]],
        "timestamps_end":   [time_to_seconds(x) for x in df["end_time"]],
        "speakers":         [f"speaker_{sid}" for sid in df["speaker_id"].astype(str).tolist()],
    })

ds = Dataset.from_list(rows).cast_column("audio", Audio(sampling_rate=16000))

split = ds.train_test_split(test_size=0.1, seed=42)
dataset = DatasetDict({"train": split["train"], "validation": split["test"]})

dataset.save_to_disk("/kaggle/working/diar_dataset")
dataset
```

## **Build and Save Full Training Dataset**

This cell rebuilds the HuggingFace dataset from the raw audio and annotation CSVs, clears any previous saved version, and saves the complete dataset to `/kaggle/working/diar_dataset` for training without creating splits.

```python
import os, shutil
import pandas as pd
from datasets import Dataset, Audio

audio_dir = "/kaggle/input/competitions/dl-sprint-4-0-bengali-speaker-diarization-challenge/diarization/diarization/train/audio"
ann_dir   = "/kaggle/input/competitions/dl-sprint-4-0-bengali-speaker-diarization-challenge/diarization/diarization/train/annotation"

OUT = "/kaggle/working/diar_dataset"
shutil.rmtree(OUT, ignore_errors=True)  

def time_to_seconds(t: str) -> float:
    # supports HH:MM:SS or HH:MM:SS.mmm
    h, m, s = t.strip().split(":")
    return int(h)*3600 + int(m)*60 + float(s)

rows = []
for wav in os.listdir(audio_dir):
    if not wav.endswith(".wav"):
        continue

    rec = wav[:-4]
    csv_path = os.path.join(ann_dir, rec + ".csv")
    if not os.path.exists(csv_path):
        continue

    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip().str.lower()

    rows.append({
        "audio": os.path.join(audio_dir, wav),
        "timestamps_start": [time_to_seconds(x) for x in df["start_time"]],
        "timestamps_end":   [time_to_seconds(x) for x in df["end_time"]],
        "speakers":         [f"speaker_{sid}" for sid in df["speaker_id"].astype(str).tolist()],
    })

dataset = Dataset.from_list(rows).cast_column("audio", Audio(sampling_rate=16000))
dataset.save_to_disk(OUT)

print("Saved:", OUT)
print("Columns:", dataset.column_names)
dataset
```
